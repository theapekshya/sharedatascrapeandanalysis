```python
from bs4 import BeautifulSoup
import requests


url = BeautifulSoup('https://merolagani.com/Floorsheet.aspx','html.parser')
soup = requests.get(url)
soup
```

```python
#soup.text

raw_code = BeautifulSoup(soup.text,'lxml')
raw_code
```

```python
table_code = raw_code.table
table_code
```

```python
tr_tags = table_code.find_all('tr')
tr_tags
```

```python
l = []
for tag in tr_tags:
    #x = tag.text
    x = tag.text.split('\n')
    if x:
        l.append(x)# append in empty list
l = [[item.strip() for item in sublist if item.strip()]for sublist in l]  #removing whitespace 
    
print(l)
```

```python
import csv
file = open('floor_data.csv','w')
y = csv.writer(file)
y.writerows(l)
file.close()
```
